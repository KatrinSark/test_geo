```python
import folium
import openrouteservice

from openrouteservice import client, places
```

```python
api_key = '5b3ce3597851110001cf62480711b9fe047e4b2ca96080257e41cd95'
clnt = client.Client(key=api_key)
sevastopol = [44.58883, 33.5224]
map1 = folium.Map(location=(sevastopol), zoom_start=12)
```

```python
params_for_city = {'profile': 'foot-walking',
              'range':[2200],
              'attributes': ['total_pop'], # Get population count for isochrones
              'locations':[[33.5224, 44.58883]]
             }
i = clnt.isochrones(**params_for_city)
```

```python
params_poi = {
    'request': 'pois',
    'geojson': i['features'][0]['geometry'],
    'filter_category_ids': [156],
    'buffer': 1
}


my_places ={}

places = clnt.places(**params_poi)[0]['features']
```

```python
for points in places:
    name = points['properties']['osm_tags']['name']
    point = points['geometry']['coordinates']
    my_places.update({f'{name}': {'location': point}})


for name, point in my_places.items():   
    folium.map.Marker(list(reversed(point['location'])), popup=name).add_to(map1)
    
map1.save("mapOutput.html")
```
